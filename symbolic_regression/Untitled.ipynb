```python
%matplotlib inline

import scipy

import numpy as np
import pandas as pd

import sklearn

from gplearn.genetic import SymbolicRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR as SVR
from sklearn.utils.random import check_random_state
from sklearn import model_selection
from sklearn.preprocessing import StandardScaler
from sklearn.neural_network import MLPRegressor

from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np
from IPython.display import Image
#import pydotplus
```

```python
# Ground truth
x0 = np.arange(-1, 1, 1/10.)
x1 = np.arange(-1, 1, 1/10.)
x0, x1 = np.meshgrid(x0, x1)
y_truth = x0**2 - x1**2 + x1 - 1

ax = plt.figure().gca(projection='3d')
ax.set_xlim(-1, 1)
ax.set_ylim(-1, 1)
surf = ax.plot_surface(x0, x1, y_truth, rstride=1, cstride=1, color='green', alpha=0.5)
plt.show()
```

```python
rng = check_random_state(0)

# Training samples
X_train = rng.uniform(-1, 1, 100).reshape(50, 2)
y_train = X_train[:, 0]**2 - X_train[:, 1]**2 + X_train[:, 1] - 1

# Testing samples
X_test = rng.uniform(-1, 1, 100).reshape(50, 2)
y_test = X_test[:, 0]**2 - X_test[:, 1]**2 + X_test[:, 1] - 1
```

```python
est_gp = SymbolicRegressor(population_size=5000,
                           generations=20, stopping_criteria=0.01,
                           p_crossover=0.7, p_subtree_mutation=0.1,
                           p_hoist_mutation=0.05, p_point_mutation=0.1,
                           max_samples=0.9, verbose=1,
                           parsimony_coefficient=0.01, random_state=0)
est_gp.fit(X_train, y_train)
```

```python
print(est_gp._program)
```

```python
est_svr = SVR()
est_svr.fit(X_train, y_train)
```

```python
y_gp = est_gp.predict(np.c_[x0.ravel(), x1.ravel()]).reshape(x0.shape)
score_gp = est_gp.score(X_test, y_test)
y_svr = est_svr.predict(np.c_[x0.ravel(), x1.ravel()]).reshape(x0.shape)
score_svr = est_svr.score(X_test, y_test)

fig = plt.figure(figsize=(12, 10))

for i, (y, score, title) in enumerate([(y_truth, None, "Ground Truth"),
                                       (y_gp, score_gp, "SymbolicRegressor"),
                                       (y_svr, score_svr, "SVR")]):

    ax = fig.add_subplot(2, 2, i+1, projection='3d')
    ax.set_xlim(-1, 1)
    ax.set_ylim(-1, 1)
    surf = ax.plot_surface(x0, x1, y, rstride=1, cstride=1, color='green', alpha=0.5)
    points = ax.scatter(X_train[:, 0], X_train[:, 1], y_train)
    if score is not None:
        score = ax.text(-.7, 1, .2, "$R^2 =\/ %.6f$" % score, 'x', fontsize=14)
    plt.title(title)

plt.show()
```

```python
def fpA(A, B, C):
    return A+5 #- 4 * B**(C * 0.1)

def fpB(A, B, C):
    return A + 8*B

def fpC(A, B, C):
    return C * 100

# synthetic environment
N = int(1e4)

A = np.random.uniform(1, 10, N)
B = np.random.uniform(1, 10, N)
C = np.random.uniform(1, 10, N)

# synthetic data
pA = fpA(A, B, C)
pB = fpB(A, B, C)
pC = fpC(A, B, C)

# combine datasets
env = pd.DataFrame({'A': A,
                    'B': B,
                    'C': C})
foram = pd.DataFrame({'pA': pA,
                      'pB': pB,
                      'pC': pC})
```

```python
tt = 1
for i in env.columns:
    for j in foram.columns:

        plt.subplot(3,3,tt)
        plt.scatter(env[i],foram[j],s=0.1,alpha=0.1)
        plt.xlabel(i)
        plt.ylabel(j)
        tt = tt +1
        
plt.tight_layout()
```

```python
# create scalers and apply to data
env_scaler = StandardScaler()
foram_scaler = StandardScaler()

foram_scaler.fit(foram)
env_scaler.fit(env)

tenv = env_scaler.transform(env)
tforam = foram_scaler.transform(foram)

# split into test/training
tenv_train, tenv_test, tforam_train, tforam_test = model_selection.train_test_split(tenv, tforam, test_size=0.3)
```

```python
nn = MLPRegressor((100, 10, 10),solver='lbfgs',activation='tanh')
nn.fit(tforam_train, tenv_train)
```

```python
def limcalc(x, pad=0.05): # simple function for plots
    rn = np.ptp(x)
    return np.nanmin(x) - rn * pad, np.nanmax(x) + rn * pad

# predictions and unscaling
train_prediction = env_scaler.inverse_transform(nn.predict(tforam_train))
test_prediction = env_scaler.inverse_transform(nn.predict(tforam_test))

fig, axs = plt.subplots(1, 3, figsize=[8, 3])

meas = env_scaler.inverse_transform(tenv_train)
pred = train_prediction

for i, (p, ax) in enumerate(zip(env.columns, axs.flat)):
    ax.set_xlabel(p)
    
    x = meas[:, i]
    y = pred[:, i]
    
    ax.scatter(x, y, alpha=0.2, s=5)
    
    mlim = limcalc(x)
    
    ax.set_xlim(mlim)
    ax.set_ylim(mlim)
    ax.plot(mlim, mlim, c='k')
    
fig.tight_layout()
```
